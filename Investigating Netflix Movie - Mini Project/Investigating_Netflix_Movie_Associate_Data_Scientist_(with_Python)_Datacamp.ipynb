```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
netflix_df
```

## What was the most frequent movie duration from year 2000 to 2010 ?

```python
# Subset the DataFrame for type "Movie"
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]
netflix_subset
```

```python
# Filter the to keep only movies released from year 2000 to 2010.
# Start by filtering out movies that were released in or after 2000
subset = netflix_subset[(netflix_subset["release_year"] >= 2000)]
subset
```

```python
# And then do the same to filter out movies released before 2010
movies_2000s = subset[(subset["release_year"] <= 2010)]
movies_2000s
```

```python
# Another way to do this step is to use the & operator which allows you to do this type of filtering in one step
movies_2000s = netflix_subset[(netflix_subset["release_year"] >= 2000) & (netflix_subset["release_year"] <= 2010)]
movies_2000s
```

```python
# Visualize the duration column of your filtered data to see the distribution of movie durations
# See which bar is the highest and save the duration value, this doesn't need to be exact!
plt.hist(movies_2000s["duration"])
plt.title('Distribution of Movie Durations from Year 2000-2010')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

## A movie is considered short if it is less than 90 minutes. Count the number of short action movies released from 2000 to 2010!

```python
# Filter the data again to keep only the Action movies
action_movies_2000s = movies_2000s[movies_2000s["genre"] == "Action"]
action_movies_2000s
```

```python
# Use a for loop and a counter to count how many short action movies there were in 2000 to 2010

# Start the counter
short_movie_count = 0

# Iterate over the labels and rows of the DataFrame and check if the duration is less than 90, if it is, add 1 to the counter, if it isn't, the counter should remain the same
for label, row in action_movies_2000s.iterrows() :
    if row["duration"] < 90 :
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count = short_movie_count

print(short_movie_count)
```

```python
# A quicker way of counting values in a column is to use .sum() on the desired column
print((action_movies_2000s["duration"] < 90).sum())
```
